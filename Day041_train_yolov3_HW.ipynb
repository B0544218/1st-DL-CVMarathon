### 訓練模型時，資料集的格式是什麼？
### VOC2007 格式所產生標註檔（*.xml）內容為PASCAL格式，主要表示物件框的方式為左上（xmin、ymin）及右下（xmax、ymax）座標，和 YOLO 格式不同，所以要依資料集清單檔（*.txt），將個別標註檔（*.xml）轉換成物件編號加上物件框中心座標及尺寸格式後再寫至另一個清單檔。
